```python
import pandas as pd
import matplotlib.pyplot as plt

# Daten aus der CSV-Datei laden
df = pd.read_csv('Migros_Flour.csv', delimiter=';')

# Preisdaten von String zu Float konvertieren (nur für Zeichenketten)
df["price_raw"] = df["price_raw"].apply(lambda x: float(str(x).replace('"', '')))

# Gewichtsdaten von String zu Float konvertieren (nur für Zeichenketten, um Probleme mit Kommas zu vermeiden)
df["weight_raw"] = df["weight_raw"].apply(lambda x: float(str(x).replace(',', '').replace('kg', '').replace('g', '')))

# Preis pro 1kg berechnen
df["price_per_kg"] = df["price_raw"] / df["weight_raw"]

# Diagramm erstellen
plt.figure(figsize=(10, 6))
plt.barh(df["productname_raw"], df["price_per_kg"])
plt.xlabel('Preis pro 1kg (in CHF)')
plt.ylabel('Produkte')
plt.title('Preis pro 1kg für verschiedene Mehlsorten')
plt.tight_layout()
plt.show()
```
